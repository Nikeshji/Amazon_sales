```python
df['Category'] = df['Category'].astype(str)
c_d=df['Category']
plt.figure(figsize=(10,6))
plt.hist(c_d,bins=15,edgecolor="Red")
plt.xticks(rotation=90)
plt.show()
```

# Most of the buyers are t-shirt

```python
#check b2b data
check_B2B=df['B2B'].value_counts()

#create pie chart
plt.pie(check_B2B, labels=check_B2B.index, autopct='%1.1f%%')
plt.show()
```

# maixmus 99.2% of buyers are retailers and 0.8% are wholesales

```python
#scatter plot
x_data =df['Category']
y_data =df['Size']

plt.scatter(x_data,y_data)
plt.xlable('Category')
plt.ylable('Size')
plt.title('Available size')
plt.show()
```

```python
top10_state= df['ship-state'].value_counts().head(10)   #for top 10 States
plt.figure(figsize=(12,6))
sns.countplot(data=df[df['ship-state'].isin(top10_state.index)],x='ship-state')
plt.xlable('State')
plt.ylable('Order_count')
plt.title('Distribution of State')
plt.xticks(rotation=90)
plt.show()
```
